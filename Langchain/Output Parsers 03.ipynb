```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage
from langchain_classic.output_parsers import DatetimeOutputParser
```

```python
chat = ChatOpenAI(
    model="gpt-4o-mini", 
    seed=365,
    temperature=0,
    max_tokens=100,
)
```

```python
parser = DatetimeOutputParser()

message_h = HumanMessage(
    content=f"""When was the Danish poet Piet Hein born?
{parser.get_format_instructions()}
"""
)
```

```python
response=chat.invoke([message_h])
response
```

```python
date_output_parser=DatetimeOutputParser()
```

```python
response_parsed=date_output_parser.invoke(response)
response_parsed
```

```python
print(date_output_parser.get_format_instructions())
```
