```python
from langchain_openai.embeddings import OpenAIEmbeddings
from langchain_chroma import Chroma
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.runnables import RunnablePassthrough
from langchain_core.runnables import RunnableParallel
from langchain_core.output_parsers import StrOutputParser
```

```python
vectorstore=Chroma(persist_directory="./intro-to-ds-lectures",embedding_function=OpenAIEmbeddings(model="text-embedding-ada-002"))
```

```python
len(vectorstore.get()["documents"])
```

```python
retriever=vectorstore.as_retriever(search_type="mmr",search_kwargs={"k":3,"lambda_mult":0.7})
```

```python
TEMPLATE = '''
Answer the following question:
{question}

To answer the question, use only the following context:
{context}

At the end of the response, specify the name of the lecture this context is taken from in the format:
Resources: *Lecture Title*
where *Lecture Title* should be substituted with the title of all resource lectures.
'''

prompt_template = PromptTemplate.from_template(TEMPLATE)
```

```python
question="what softwares do data scientist use?"
```

```python
chat = ChatOpenAI(model= 'gpt-4', 
                  seed=365,
                  max_tokens = 250)
```

```python
chain = ({'context': retriever, 
         'question': RunnablePassthrough()} 
         | prompt_template 
         | chat 
         | StrOutputParser())
```

```python
chain.invoke(question)
```
