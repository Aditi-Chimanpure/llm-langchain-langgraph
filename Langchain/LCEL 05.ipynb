```python
%load_ext dotenv
%dotenv
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from langchain_core.prompts import ChatPromptTemplate
```

```python
RunnablePassthrough().invoke("hi")
```

```python
chat_template_tools=ChatPromptTemplate.from_template("""what are the five most imp tools a {job_title} needs?
Anser only by listing the tools""")
chat_template_tools
```

```python
chat_template_strategy=ChatPromptTemplate.from_template('''
Considering the tools provided, develop a strategy for effectively learning and mastering them:
{tools}
''')
```

```python
chat = ChatOpenAI(model_name = 'gpt-4', 
                  model_kwargs = {'seed':365},
                  temperature = 0,
                  max_tokens = 100)
```

```python
string_parser=StrOutputParser()
```

```python
chain_tools=chat_template_tools|chat|string_parser|{"tools":RunnablePassthrough()}
```

```python
chain_strategy=chat_template_strategy|chat|string_parser
```

```python
print(chain_tools.invoke({"job_title":"ai engineer"}))
```

```python
print(chain_strategy.invoke({"tools":"""1. Python Programming Language
2. TensorFlow
3. Jupyter Notebook
4. Keras
5. PyTorch"""}))
```

```python
chain_combined=chain_tools|chain_strategy
```

```python
print(chain_combined.invoke({"job_title":"python engineer"}))
```

```python
chain_combined2=(chat_template_tools|chat|string_parser|{"tools":RunnablePassthrough()}|
                 chat_template_strategy
                 |chat|string_parser)
```

```python
print(chain_combined2.invoke({"job_title":"backend engineer"}))
```
