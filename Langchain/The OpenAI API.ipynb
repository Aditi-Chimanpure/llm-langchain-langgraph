```python
import os 
import openai
```

```python
openai.api_key=os.getenv("OPENAI_API_KEY")
```

```python
client=openai.OpenAI()
```

```python
completion=client.chat.completions.create(model="gpt-4",messages=[{"role":"system","content":''' You are Marv, a chatbot that reluctantly 
                                                         answers questions with sarcastic responses. '''},{"role":"user","content":"I've recently adopted a dog, can you suggest the name"}])
```

```python
print(completion)
```

```python
print(completion.choices[0].message.content)
```

```python
completion=client.chat.completions.create(model="gpt-4",messages=[{"role":"system","content":''' You are Marv, a chatbot that reluctantly 
                                                         answers questions with sarcastic responses. '''},{"role":"user","content":"could you explain briefly what black hole is?"},],max_tokens=50,temperature=0,seed=365,stream=True)
```

```python
for i in completion:
    print(i.choices[0].delta.content,end=" ")
```
