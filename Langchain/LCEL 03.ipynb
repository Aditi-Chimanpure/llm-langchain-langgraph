```python
%load_ext dotenv
%dotenv
```

```python
from langchain_openai import ChatOpenAI

from langchain_core.prompts import ChatPromptTemplate
```

```python
chat_template=ChatPromptTemplate.from_messages([('human', 
     "I've recently adopted a {pet} which is {breed} . Could you suggest training tips \n" )])
```

```python
chat = ChatOpenAI(model_name = 'gpt-4', 
                  model_kwargs = {'seed':365},
                  temperature = 0,
                  max_tokens = 100)
```

```python
chain=chat_template|chat
```

```python
chain.invoke({"pet":"dog","breed":"husky"})
```

```python
%%time
chain.batch([{"pet":"dog","breed":"husky"},{"pet":"dragon","breed":"night fury"}])
```

```python
%%time
chain.invoke({"pet":"dog","breed":"husky"})
```

```python
%%time
chain.invoke({"pet":"dragon","breed":"night fury"})
```

```python
response=chain.stream({"pet":"dragon","breed":"night fury"})
```
