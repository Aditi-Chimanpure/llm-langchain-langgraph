### create indexes

```python
import pinecone
from pinecone import Pinecone, ServerlessSpec
import os
from dotenv import load_dotenv, find_dotenv
```

```python
load_dotenv(find_dotenv(),override=True)
```

```python
pc=Pinecone(api_key=os.environ.get("PINECONE_API_KEY"))
```

```python
pc.list_indexes()
```

```python
index_name = "my-index"
dimension = 3
metric = "cosine"
```

```python
if index_name in [index.name for index in pc.list_indexes()]:
    pc.delete_index(index_name)
    print(f"{index_name} succesfully deleted.")
else:
     print(f"{index_name} not in index list.")
```

```python
pc.list_indexes()
```

```python
pc.create_index(name=index_name,dimension=dimension,metric=metric,spec=ServerlessSpec(cloud="aws",region="us-east-1"))
```

```python
pc.list_indexes()
```

```python
pc.create_index(name='my-index-2',dimension=1024,metric="euclidean",spec=ServerlessSpec(cloud="aws",region="us-east-1"))
```
