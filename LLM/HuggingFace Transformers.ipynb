```python
from transformers import AutoTokenizer
import torch
```

```python
model="bert-base-uncased"
```

```python
tokenizer=AutoTokenizer.from_pretrained(model)
```

```python
sentence="I'm so excited to be learnign about large language model"
```

```python
input_ids=tokenizer(sentence)
print(input_ids)
```

```python
tokens=tokenizer.tokenize(sentence)
print(tokens)
```

```python
token_ids=tokenizer.convert_tokens_to_ids(tokens)
print(token_ids)
```

```python
decoded_ids=tokenizer.decode(token_ids)
print(decoded_ids)
```

```python
from transformers import AutoTokenizer,AutoModelForSequenceClassification
import torch
```

```python
print(sentence)
print(input_ids)
```

```python
tokenizer=AutoTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

```python
input_ids_pt=tokenizer(sentence,return_tensors="pt")
print(input_ids_pt)
```

```python
model=AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

```python
print(input_ids_pt["input_ids"].shape)
```

```python
print(model.name_or_path)
```

```python
import torch

device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")

print("Before model")

filtered_inputs = {
    k: v.to(device) 
    for k, v in input_ids_pt.items() 
    if k != "token_type_ids"
}

model = model.to(device)

with torch.no_grad():
    logits = model(**filtered_inputs).logits

print("After model")

predicted_class_id = logits.argmax().item()
print(model.config.id2label[predicted_class_id])
```
