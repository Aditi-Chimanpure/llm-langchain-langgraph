```python
def generate_text(prompt):
    response=client.responses.create(model="gpt-4.1-mini",input=prompt,max_output_tokens=16,
                                    temperature=0.4)
    return response.output_text.strip()
```

```python
def poetic_chatbot(prompt):
    response = client.responses.create(
        model="gpt-4.1-mini",
        input = [
            {
                "role": "system",
                "content": "You are a poetic chatbot."
            },
            {
                "role": "user",
                "content": "When was Google founded?"
            },
            {
                "role": "assistant",
                "content": "In the late '90s, a spark did ignite, Google emerged, a radiant light. By Larry and Sergey, in '98, it was born, a search engine new, on the web it was sworn."
            },
            {
                "role": "user",
                "content": "Which country has the youngest president?"
            },
            {
                "role": "assistant",
                "content": "Ah, the pursuit of youth in politics, a theme we explore. In Austria, Sebastian Kurz did implore, at the age of 31, his journey did begin, leading with vigor, in a world filled with din."
            },
            {
                "role": "user",
                "content": prompt
            }
        ],
        temperature = 1,
        
    )
    return response.output_text.strip()
```

```python
from langchain_community.document_loaders import WebBaseLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain_community.vectorstores import FAISS
```

```python
url = "https://edition.cnn.com/"
```

```python
loader=WebBaseLoader(url)
```

```python
raw_documents=loader.load()
```

```python
text_splitter= RecursiveCharacterTextSplitter()
```

```python
documents=text_splitter.split_documents(raw_documents)
```

```python
embedding=OpenAIEmbeddings(openai_api_key = config.api_key)
```

```python
vectorstore = FAISS.from_documents(documents, embedding)
```

```python
retriever = vectorstore.as_retriever()
```

```python
llm = ChatOpenAI(
    openai_api_key=config.api_key,
    model="gpt-4.1-mini",
    temperature=0
)
```

```python
chat_history = []
```

```python
query="what is the latest news?"
```

```python
relevant_docs=retriever.invoke(query)
```

```python
context = "\n\n".join(doc.page_content for doc in relevant_docs)
```

```python
history_text = "\n".join(
    f"User: {q}\nAssistant: {a}" for q, a in chat_history
)
```

```python
prompt = f"""
Use the context below to answer the question.

Conversation history:
{history_text}

Context:
{context}

Question:
{query}
"""

response = llm.invoke(prompt)
```

```python
response = llm.invoke(prompt)
chat_history.append((query, response.content))
```
